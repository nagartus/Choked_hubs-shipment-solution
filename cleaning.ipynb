```python
import pandas as pd

# Replace 'your_file.csv' with the actual file path
file_path = 'finalsorted.csv'

try:
    # Try reading the CSV file with 'utf-8' encoding
    df = pd.read_csv(file_path, encoding='utf-8')
except UnicodeDecodeError:
    # If an error occurs, try 'latin-1' encoding
    df = pd.read_csv(file_path, encoding='latin-1')

# Now you can work with the DataFrame 'df'
df
```

```python
import pandas as pd

# Assuming df is your DataFrame
# Sort the DataFrame by 'shipment_id' and 'ctime'
df.sort_values(by=['shipment_id', 'ctime'], inplace=True)

# Identify the last row for each 'shipment_id'
last_rows = df.groupby('shipment_id').tail(1)

# Fill the 'location' column with 'latest_location' where 'location' is null or blank
mask = (last_rows['location'].isnull()) | (last_rows['location'] == '')
last_rows.loc[mask, 'location'] = last_rows['latest_location']

# Merge the updated last_rows back into the original DataFrame
df = pd.merge(df, last_rows[['shipment_id', 'ctime', 'location']], on=['shipment_id', 'ctime'], how='left', suffixes=('', '_update'))

# Fill missing values in the 'location' column with the updated values
df['location'].fillna(df['location_update'], inplace=True)
df.drop(columns=['location_update'], inplace=True)

# Display the updated DataFrame
print(df)
```

```python
import pandas as pd

# Assuming df is your DataFrame sorted by 'shipment_id' and 'ctime'
df.sort_values(by=['shipment_id', 'ctime'], inplace=True)

# Identify the last row for each 'shipment_id'
last_rows = df.groupby('shipment_id').tail(1)

# Check if 'location' is null or blank, 'latest_location' is not empty, and 'latest_location' is not null or blank
filtered_rows = last_rows[
    (last_rows['location'].isnull() | (last_rows['location'] == '')) & 
    (last_rows['latest_location'].notna() & (last_rows['latest_location'] != ''))
]

# Display the filtered rows
print(filtered_rows)
```

```python
import pandas as pd

# Assuming df is your DataFrame
# Filter rows where 'location' is null or blank
rows_to_delete = df[(df['location'].isnull() | (df['location'] == '')) & 
                    (df['latest_location'].notna() & (df['latest_location'] != ''))]

# Drop the rows from the original DataFrame
df = df.drop(rows_to_delete.index)

# Display the updated DataFrame
print(df)
```
